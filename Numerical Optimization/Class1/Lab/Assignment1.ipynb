```python
samples_10000 = np.random.uniform(-1, 1, 10000)
samples_100000 = np.random.uniform(-1, 1, 100000)

plt.figure(figsize=(12, 6))

#bins = 100 corresponds to number of intervals
#density = true normalizes the histogram so that the area under the histogram equals 1, making it a probability density.
plt.subplot(1, 2, 1)
plt.hist(samples_10000, bins=100, density=True, alpha=0.6, color='b', label='Histogram (10,000 samples)')
x = np.linspace(-1, 1, 1000)
plt.plot(x, np.ones_like(x) / 2, 'r-', lw=2, label='PDF')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Histogram and PDF for 10,000 samples')
plt.legend()

#pdf is a horizontal line at a height of 1/(b-a) a and b are lower and upper bounds
plt.subplot(1, 2, 2)
plt.hist(samples_100000, bins=100, density=True, alpha=0.6, color='g', label='Histogram (100,000 samples)')
plt.plot(x, np.ones_like(x) / 2, 'r-', lw=2, label='PDF')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Histogram and PDF for 100,000 samples')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#loc specifies mean and scale specifies standard deviation as given N(5,2) and N(3,1)

samples_10000 = np.random.normal(loc=[5, 3], scale=[2, 1], size=(10000, 2))
samples_100000 = np.random.normal(loc=[5, 3], scale=[2, 1], size=(100000, 2))

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(samples_10000[:, 0], samples_10000[:, 1], 'b.', markersize=2, label='Generated Points (10,000 samples)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Generated Points (10,000 samples)')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(samples_100000[:, 0], samples_100000[:, 1], 'g.', markersize=2, label='Generated Points (100,000 samples)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Generated Points (100,000 samples)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python

#The Euclidean length of a vector is computed by taking the square root of the sum of the squares of its elements
def euclidean_length(x):
    return np.sqrt(np.sum(x**2))

#The weighted average is calculated by multiplying each element of x with its corresponding weight from w, 
# summing up these products, and then dividing by the sum of the weights.
def weighted_average(x, w):
    return np.sum(x * w) / np.sum(w)

#The Euclidean distance is the straight-line distance between two points in Euclidean space
def euclidean_distance(x, y):
    return np.sqrt(np.sum((x - y)**2))

#The dot product is the sum of the products of the corresponding elements of the two vectors
def dot_product(x, y):
    return np.sum(x * y)

x = np.array([1, 2, 3])
length_x = euclidean_length(x)

w = np.array([0.1, 0.2, 0.3])
weighted_avg = weighted_average(x, w)

y = np.array([4, 5, 6])
distance_xy = euclidean_distance(x, y)

dot_xy = dot_product(x, y)

print("Dot product of vectors x and y:", dot_xy)
print("Euclidean distance between vectors x and y:", distance_xy)
print("Weighted average of vector x with weights w:", weighted_avg)
print("Euclidean length of vector x:", length_x)
```

```python
iris = load_iris()

data = iris.data
target = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

f = plt.figure(figsize=(8, 6))

for i in range(len(target_names)):
    plt.scatter(data[target == i, 0], data[target == i, 1], s=100, c="r", marker=".", label=target_names[i])

plt.xlabel('sepal width (cm)')
plt.ylabel('sepal length (cm)')
plt.title("Iris")
plt.show()
```
